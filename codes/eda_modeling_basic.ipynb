```python
def check_outlier(data, column): 

    # lower, upper 글로벌 변수 선언하기     
    global lower, upper    
    
    # 4분위수 기준 지정하기     
    q25, q75 = np.quantile(data.iloc[:, column], 0.25), np.quantile(data.iloc[:, column], 0.75)          
    
    # IQR 계산하기     
    iqr = q75 - q25    
    
    # outlier cutoff 계산하기     
    cut_off = iqr * 1.5          
    
    # lower와 upper bound 값 구하기     
    lower, upper = q25 - cut_off, q75 + cut_off      

    return lower, upper
```

```python
# 이상치 영역을 나타내는 그래프 그리기 - 랜덤하게 일부 컬럼만 선별하여

fig, (ax1, ax2, ax3, ax4)= plt.subplots(1, 4, figsize=(32, 6))

fig.suptitle('Random histogram with outliers', fontsize=16)

rc1 = np.random.randint(1, len(train_df.columns))
lower, upper = check_outlier(train_df, rc1)

# 컬럼별 데이터 분포 확인하기
sns.distplot(train_df.iloc[:, rc1], color = 'y', ax = ax1)

# 이상치 영역 박스 그리기
ax1.axvspan(lower, train_df.iloc[:, rc1].min(), alpha=0.2, color='c')
ax1.axvspan(upper, train_df.iloc[:, rc1].max(), alpha=0.2, color='c')
ax1.set_title(train_df.columns[rc1])


rc2= np.random.randint(1, len(train_df.columns))
lower, upper = check_outlier(train_df, rc2)

sns.distplot(train_df.iloc[:, rc2], color = 'y', ax = ax2)

ax2.axvspan(lower, train_df.iloc[:, rc2].min(), alpha=0.2, color='c')
ax2.axvspan(upper, train_df.iloc[:, rc2].max(), alpha=0.2, color='c')
ax2.set_title(train_df.columns[rc2])


rc3 = np.random.randint(1, len(train_df.columns))
lower, upper = check_outlier(train_df, rc3)

sns.distplot(train_df.iloc[:, rc3], color = 'y', ax = ax3)

ax3.axvspan(lower, train_df.iloc[:, rc3].min(), alpha=0.2, color='c')
ax3.axvspan(upper, train_df.iloc[:, rc3].max(), alpha=0.2, color='c')
ax3.set_title(train_df.columns[rc3])


rc4 = np.random.randint(1, len(train_df.columns))
lower, upper = check_outlier(train_df, rc4)

sns.distplot(train_df.iloc[:, rc4], color = 'y', ax = ax4)

ax4.axvspan(lower, train_df.iloc[:, rc4].min(), alpha=0.2, color='c')
ax4.axvspan(upper, train_df.iloc[:, rc4].max(), alpha=0.2, color='c')
ax4.set_title(train_df.columns[rc4])

plt.show()

    
# 특정 컬럼들에서 이상치가 넓은 면적으로 보이는 경우 삭제 후 모델링 진행이 필요해보임. 
```

# 모델링 준비하기

```python
# 독립, 종속 변수 데이터로 분리하기
train_x = train_df[train_df.columns[:len(x_features)]]
train_y = train_df[train_df.columns[len(x_features):]]
```

```python
import random
def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
```

```python
# from sklearn.preprocessing import StandardScaler

# ss_scale = StandardScaler()
# train_x = ss_scale.fit_transform(train_x)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(train_x, train_y, test_size=0.2, random_state=seed_everything(2022), shuffle=True)
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

## 모델링 진행 - Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.multioutput import MultiOutputRegressor

seed_everything(2022)
lr_model = LinearRegression()
LR = MultiOutputRegressor(lr_model).fit(train_x, train_y)
print('Done.')
```

```python
LR
```

```python
print(f'R2 score of LR (training): {round(LR.score(train_x, train_y), 3)}')
print(f'R2 score of LR (validation): {round(LR.score(X_val, y_val), 3)}')
```

```python
# 상위 검증데이터 50개 데이터에 대하여 그래프 그려서 확인하기 (분포 확인)
f = plt.figure(figsize=(20, 20))
f.add_subplot(2,1,1)
plt.title("Xs validation data", fontsize=16)
plt.plot(X_val[:50])
plt.xlabel(None)
plt.xticks(rotation=45)
plt.legend(y_features)

f.add_subplot(2,1,2)
plt.title("Ys validation data", fontsize=16)
plt.plot(y_val[:50])
plt.xlabel(None)
plt.xticks(rotation=45)
plt.legend(y_features)

plt.show()
```

```python
# test_x = ss_scale.fit_transform(test_df)
preds = LR.predict(X_val)
print('Done.')
```

```python
preds[0]
```

```python
from sklearn.metrics import mean_squared_error
print("y MSE:%.4f" % mean_squared_error(y_val, preds))
```

```python
# 검증데이터에 대한 예측 결과  그래프를 통해 확인하기 - 첫번째 컬럼인 안테나 Gain 평균 (각도1)에 대하여 확인하기

plt.figure(figsize=(15, 10))
x_ax = range(len(X_val))
plt.plot(x_ax, y_val.iloc[:,0], label="y1-test", color='c') 
plt.plot(x_ax, preds[:,0], label="y1-pred", color='b')
plt.title("검증데이터와 예측데이터", fontsize=16)
plt.legend()
plt.show()
```

# 결과 정리
- 일부 컬럼들의 경우 이상치의 영향력이 크기 때문에 해당 영향력을 삭제할 필요가 있음. (이상치 탐지 - isolation forest 활용)
- 전처리를 하지 않은 상태에서는 독립, 종속 변수 간의 영향력보다 독립변수는 독립변수끼리, 종속변수는 종속변수끼리의 영향력을 미치는 것으로 보여짐
- 과적합, 다중공선성 등의 문제들의 영향력을 무시할 수 없기 때문에 전처리가 제대로 해주어야 함.
- 변수 별로 스케일링을 별도로 진행해줄 필요가 있음.

=> 공정 관련 데이터의 경우 불량률, 이상치 등이 데이터에 속해있으므로 해당 부분에 대한 전처리를 진행해준 다음 모델링을 진행
